```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "名前":["大沢","河合","上田"],
    "入社日":['1995-04-01','1998-07-01','2000-01-01'],
    "所属":["海外営業部","本部人事部","広島営業所"]
})
display(df1)
```
